```python
import numpy as np
import networkx as nx
%matplotlib inline

from matplotlib import pyplot as plt
plt.style.use('ggplot')
```

```python
reads = ["ATG", 
         "TGG", 
         "GGC", 
         "GCG", 
         "CGT", 
         "GTG", 
         "TGC", 
         "GCA", 
         "CAA", 
         "AAT"]
nodos = ["AT", "TG", "GG", "GC", "CG", "GT", "CA", "AA"]

aristas = [ ("AT", "TG"),  
            ("TG", "GG"), 
            ("GG", "GC"), 
            ("GC", "CG"), 
            ("CG", "GT"), 
            ("GT", "TG"),
            ("TG", "GC"), 
            ("GC", "CA"), 
            ("CA", "AA"), 
            ("AA", "AT")]

dicedgs = { ("AT", "TG"):'ATG',  
            ("TG", "GG"):"TGG", 
            ("GG", "GC"):"GGC", 
            ("GC", "CG"):"GCG", 
            ("CG", "GT"):"CGT", 
            ("GT", "TG"):"GTG",
            ("TG", "GC"):"TGC", 
            ("GC", "CA"):"GCA", 
            ("CA", "AA"):"CAA", 
            ("AA", "AT"):"AAT"}
```

```python
G = nx.DiGraph()
```

```python
G.add_nodes_from(nodos)
G.nodes()
```

```python
G.add_edges_from(aristas)
G.edges()
```

```python
fig = plt.figure(figsize=(10,10))
pos=nx.circular_layout(G)
nx.draw_networkx_nodes(G, pos, node_size=500)
nx.draw_networkx_edges(G, pos, edgelist=aristas, width=[5]*len(aristas))
nx.draw_networkx_labels(G, pos, font_size=20, font_family="sans-serif")
nx.draw_networkx_edge_labels(G, pos, edge_labels=dicedgs, font_size=20, font_family="sans-serif")
```
