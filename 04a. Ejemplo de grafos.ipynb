# Teoría de grafos



Para representar grafos en la computadora hay varias opciones, desde las implementaciones propias hasta las bibliotecas especializadas, entre ellas está Networkx.

```python
import pandas   as pd
import networkx as nx
import numpy    as np
import matplotlib
%matplotlib inline

from matplotlib import pyplot as plt
```

```python
G = nx.Graph()
```

```python
G.add_node('a')
```

```python
G.nodes()
```

```python
nodos = ['a','b','c','d','e']
G.add_nodes_from( nodos )
```

```python
G.add_edges_from([('a','b'),
                  ('a','e'),
                  ('a','d'),
                  ('c','b'),
                  ('c','e'),
                  ('d','e')])
```

```python
G.edges()
```

```python
fig = plt.figure(figsize=(10,10))
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, node_size=500, node_color='yellow')
nx.draw_networkx_edges(G, pos, edgelist=G.edges(), width=10)
nx.draw_networkx_labels(G, pos, font_size=20, font_family="sans-serif")
plt.axis("off")
plt.show()
```
